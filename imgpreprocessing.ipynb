```python
import cv2

import numpy as np
from google.colab.patches import cv2_imshow
```

```python
#read the image
img = cv2.imread('/content/fuaads_car.jpeg')
cv2_imshow(img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
print(img)
```

```python
#resize the image
resized_image = cv2.resize(img, (400, 300))
cv2_imshow(resized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#grayshading the image to simplfy processing step
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

```python
#blurred the image
blurred_image = cv2.GaussianBlur(img, (5, 5), 0)

cv2_imshow(blurred_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#normalize the image
normalized_image = cv2.normalize(blurred_image, None, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX)
cv2_imshow(normalized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
denoised = cv2.fastNlMeansDenoising(img,None, h=10)
cv2_imshow(blurred_image)
cv2_imshow(denoised)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
alpha = 1.5
adj= cv2.convertScaleAbs(blurred_image,alpha=alpha,beta=0)
cv2_imshow(blurred_image)
cv2_imshow(adj)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
